DIABETES DETECTION USING NEURAL NETWORKS


```python
import sklearn
```

```python
import tensorflow as tf
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

```

```python
diabetes = pd.read_csv('diabetes.csv')
```

```python
diabetes.head(7) #shows first seven rows. Passing no arguements gives us five rows.
```

```python
diabetes.tail(2) #gives us last two rows of the dataset
```

```python
diabetes.info() #gives us the info of the dataset 
diabetes.describe() #gives us description of the datavalues
```

```python
#splitting the data set

X = diabetes.iloc[:,0:8]
y = diabetes.iloc[:,8]
print(X)
print(y)
```

```python
#scaling operation peformed. It is a must in ANN
from sklearn.preprocessing import StandardScaler
sc = StandardScaler(); #creating an object of the StandardScaler class
X = sc.fit_transform(X);


#splitting the X values into training and test data
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X , y , test_size = 0.2)


```

**BUILDING THE NEURAL NETWORK**

```python
nn = tf.keras.models.Sequential()  #we create a sequential model of neural network
#adding layers
nn.add(tf.keras.layers.Dense(units = 400 , activation = 'relu'))  #dense layer 1
nn.add(tf.keras.layers.Dropout(0.2)) # dropout
nn.add(tf.keras.layers.Dense(units = 400 , activation = 'relu')) #dense layer 2 
nn.add(tf.keras.layers.Dropout(0.2)) #dropout 
nn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))


```

```python
nn.compile(optimizer = 'Adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])
```

```python
#keeping track of model performance 
epochs_hist = nn.fit(X_train , y_train , epochs = 200)
```

```python
y_pred = nn.predict(X_test)
```
